### ðŸ“Š CO RELATION BETWEEN NUMERIC VALUE IN DATA

```python
c = df.corr(numeric_only=True)
c
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(c, annot=True)
plt.title('CO-RELATION BETWEEN NUMERICAL VALUES', size=20)
plt.show()
```

## âœ’ï¸ Inference:
The diagonal of the matrix is all 1s, as a variable is always perfectly correlated with itself. 
The off-diagonal values are very close to zero, suggesting that there is no strong linear relationship between any of the numerical variables in this dataset. The highest correlation seems to be between Year and Engine_Size_L, which is a weak negative correlation, and Mileage_KM and Year, which is a weak positive correlation.
This heatmap is an excellent way to quickly visualize and understand the relationships between multiple numerical variables in a dataset.
